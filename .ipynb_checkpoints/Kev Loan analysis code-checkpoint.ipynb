```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#plotting data in the graph
Loan_type = ['type1', 'type2', 'type3']
Male_type = [31591,5701,5054]
Female_type = [21847,4544,875]
Joint_type = [31762,4926,4711]
Not_available = [27973,5591,4095 ]

# create an array for the x-axis positions
             
x = np.arange(len(Loan_type))
             
#set the with of the bars
bar_width = 0.2
             
# create a bar graph for each product
plt.bar(x - bar_width,Male_type, width=bar_width, label='Male', color='b' )
plt.bar(x,Female_type, width=bar_width, label='Female', color='g')
plt.bar(x +bar_width, Joint_type, width=bar_width, label='Joint', color='r')
plt.bar(x + 2*bar_width, Not_available, width=bar_width, label='No_Sex', color='y')
        
        
# Add labels and title
plt.xlabel('Loan Type')
plt.ylabel('Gender Type')
plt.title('Comparison: Gender and loan loan type ')
plt.xticks(x, Loan_type)  # Set x-axis labels

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
